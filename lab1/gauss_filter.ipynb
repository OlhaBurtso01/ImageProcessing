```python
import math
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_cv_img(input_image, output_image):
    """
    Converts an image from BGR to RGB and plots
    """
    fig, ax = plt.subplots(nrows=1, ncols=2)
    ax[0].imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
    ax[0].set_title('Input Image')
    ax[0].axis('off')
    ax[1].imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
    ax[1].set_title('Gaussian Blurred')
    ax[1].axis('off')
    plt.show()
```

```python
def kernel(size, k, sigma):
    kernel_matrix = np.zeros((size, size), np.float32)
    for i in range(size):
        for j in range(size):
            norm = math.pow(i - k, 2) + math.pow(j - k, 2)
            kernel_matrix[i, j] = math.exp(-norm / (2 * math.pow(sigma, 2))) / 2 * math.pi * pow(sigma, 2)
    return kernel_matrix / np.sum(kernel_matrix)
```

```python
def mygaussFilter(imgg, kernel):
    ih, iw = imgg.shape
    kh, kw = kernel.shape
    for i in range(int(kh / 2), ih - int(kh / 2)):
        for j in range(int(kh / 2), iw - int(kh / 2)):
            suma = 0
            for m in range(0, kh):
                for l in range(0, kh):
                    suma += imgg[i - int(kh / 2) + m, j - int(kh / 2) + l] * kernel[m, l]
            imgg[i, j] = suma
    return imgg
```

```python
img_org = cv2.imread("images\medium.jpg")
img = img_org.copy()
k = 1
size = 2 * k + 1
sigma = 1

img_gauss_B = mygaussFilter(img[:, :, 0], kernel(size, k, sigma))
img_gauss_G = mygaussFilter(img[:, :, 1], kernel(size, k, sigma))
img_gauss_R = mygaussFilter(img[:, :, 2], kernel(size, k, sigma))
img_gauss = np.dstack([img_gauss_B, img_gauss_G, img_gauss_R])

plot_cv_img(img_org, img_gauss)
```
